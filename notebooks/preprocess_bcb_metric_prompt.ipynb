# Preprocess bcb prompts and metrics

Prompts and metrics are all saved in `data/bcb.csv` (or `metric/bcb.csv`).

```python
import pandas as pd

file_path_list = ['../data/bcb_funcs_mod.csv']

# Load the CSV file
df = pd.read_csv(file_path_list[0])

# Display the first few rows of the DataFrame
df.head()
```

```python
import os
import pandas as pd
import json

task_type_list = ['bcb']
# Create the folder to save the modified prompts
folder_path = '../data/'
os.makedirs(folder_path, exist_ok=True)
for file_path, task_type in zip(file_path_list, task_type_list):
    df = pd.read_csv(file_path)
    # Iterate over the DataFrame
    for index, row in df.iterrows():
        prompt_dict = {}

        prompt_dict['prompt'] = row['Prompt']
        prompt_dict['data_source_type'] = '1_bcb'
        # Create the folder for each prompt
        prompt_folder = os.path.join(folder_path, task_type + str(row['Task ID']).split('/')[-1])
        os.makedirs(prompt_folder, exist_ok=True)
        
        # Save the prompt in a json file
        prompt_file = os.path.join(prompt_folder, f"prompt.json")
        with open(prompt_file, 'w') as file:
            json.dump(prompt_dict, file)
```

```python
import csv
import os
import yaml

# Path to the CSV file
file_path_list = ['../metric/bcb_funcs_mod.csv']
task_type_list = ['bcb']
# csv_file = '../metric/csv_excel_prompt_metric.csv'

# Create the metric folder if it doesn't exist
metric_folder = '../metric/'
os.makedirs(metric_folder, exist_ok=True)

for csv_file, task_type in zip(file_path_list, task_type_list):
    # Read the CSV file
    with open(csv_file, 'r') as file:
        reader = csv.reader(file)
        header = next(reader)  # Get the header row

        # Process each line in the CSV file
        for i, line in enumerate(reader):
            # print(i)
            # Create a folder for each line
            line_folder = os.path.join(metric_folder, f"{task_type}{str(line[0]).split('/')[-1]}")
            os.makedirs(line_folder, exist_ok=True)

            # Create a dictionary for the line
            line_dict = {'prompt': line[1], 'TMC-list': [], 'testcase': line[2], 'ground_truth': line[3], 'data': line[4]}
            for j in range(5, len(header), 3):
                task = line[j]
                metric = line[j+1]
                function = line[j+2]
                if task and metric and function:  # Check if all three values are non-empty
                    tmc_dict = {'task_name': task, 'metric': metric, 'code': function}
                    line_dict['TMC-list'].append(tmc_dict)

            # Save the line dictionary as a YAML file
            yaml_file = os.path.join(line_folder, 'metric.yaml')
            with open(yaml_file, 'w') as yaml_file:
                yaml.dump(line_dict, yaml_file)
```
