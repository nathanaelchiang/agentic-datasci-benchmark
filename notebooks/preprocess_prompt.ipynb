```python
import os
import pandas as pd
import json

task_type_list = ['csv_excel_', 'dl_', 'human_']
# Create the folder to save the modified prompts
folder_path = '../data/'
os.makedirs(folder_path, exist_ok=True)
for file_path, task_type in zip(file_path_list, task_type_list):
    df = pd.read_csv(file_path)
    # Iterate over the DataFrame
    for index, row in df.iterrows():
        prompt_dict = {}
        use_orig = False if row['Qualified (Y/N)'] == 'N' else True
        if use_orig:
            prompt = row['Prompt']
        else:
            prompt = row['Modified Prompt (If not qualified)']
        data_source_type = row['data source type']
        prompt_dict['prompt'] = prompt
        prompt_dict['data_source_type'] = data_source_type
        # Create the folder for each prompt
        prompt_folder = os.path.join(folder_path, task_type + str(row['Unnamed: 0']))
        os.makedirs(prompt_folder, exist_ok=True)
        
        # Save the prompt in a json file
        prompt_file = os.path.join(prompt_folder, f"prompt.json")
        with open(prompt_file, 'w') as file:
            json.dump(prompt_dict, file)
```
